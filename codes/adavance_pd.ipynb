```python
import pandas as pd

# Customers dataset
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'name': ['Pravin', 'Kishore', 'Farhan', 'Rabi'],
    'region': ['Delhi', 'Madurai', 'Kolkata', 'Mumbai']
})
```

```python
orders = pd.DataFrame({
    'order_id': [101, 102, 103, 104, 105],
    'customer_id': [1, 2, 2, 3, 5],  # Note: customer_id=5 doesn't exist in customers
    'product_id': ['A', 'B', 'C', 'A', 'B'],
    'amount': [250, 120, 300, 400, 500]
})
```

```python
# Products dataset
products = pd.DataFrame({
    'product_id': ['A', 'B', 'C'],
    'product_name': ['Laptop', 'Phone', 'Tablet'],
    'category': ['Electronics', 'Electronics', 'Electronics']
})
```

```python
# Imagine monthly sales data

jan_sales = pd.DataFrame({
    'order_id': [201, 202],
    'amount': [150, 200]
})

feb_sales = pd.DataFrame({
    'order_id': [203, 204],
    'amount': [300, 400]
})
```

```python
# Imagine monthly sales data

jan_sales = pd.DataFrame({
    'order_id': [201, 202],
    'amount': [150, 200]
})

feb_sales = pd.DataFrame({
    'order_id': [203, 204],
    'amount': [300, 400]
})

# Concatenate the monthly data
all_sales = pd.concat([jan_sales, feb_sales], ignore_index=True)

print(all_sales)
```

```python
# Merge Customers + Orders

# Left join: keep all customers, attach orders if available
cust_orders = pd.merge(customers, orders,
                       on='customer_id',
                       how='left')

print(cust_orders)
```

```python
# Add product details to orders

orders_products = pd.merge(orders, products,
                           on='product_id',
                           how='left')

print(orders_products)
```

```python
# Join (using index)

# Set index for join
customers_indexed = customers.set_index('customer_id')
orders_indexed = orders.set_index('customer_id')

# Join the datasets
joined_data = customers_indexed.join(orders_indexed, how='inner')

print(joined_data)
```
