```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\testing\BCT_3\data\query.csv")
```

```python
df.head()
```

```python
sns.histplot(df["mag"], bins=20)
plt.title("Magnitude Distribution in India Region")
plt.show()
```

```python
df["time"] = pd.to_datetime(df["time"])
df = df.sort_values("time")

plt.plot(df["time"], df["mag"])
plt.title("Earthquake Trend (Last 6 Months)")
plt.show()
```

```python
df["depth_type"] = df["depth"].apply(lambda x: "Shallow (<70km)" if x < 70 else "Deep")
df["depth_type"].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Earthquake Depth Distribution")
plt.show()
```
