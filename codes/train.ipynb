```python
import pandas as pd 
import seaborn as sns
```

```python
# Load Titanic dataset
df = sns.load_dataset("titanic")
```

```python
print(df)
```

```python
print(df.columns)
```

```python
df_renamed = df.rename(columns={
    "age": "Age",
    "sex": "Gender",
    "class": "PassengerClass"
})

print(df_renamed.head())
```

```python
df['age'] = df['age'].astype(int, errors='ignore')
```

```python
# Group by gender and calculate average age
avg_age_by_gender = df.groupby("sex")["age"].mean()

print(avg_age_by_gender)
```

```python
# Group by passenger class and calculate survival rate
survival_rate_by_class = df.groupby("class")["survived"].mean()

print(survival_rate_by_class)
```

```python
# Group by multiple columns
survival_rate = df.groupby(["sex", "class"])["survived"].mean()

print(survival_rate)
```

```python
# Aggregate multiple statistics
agg_stats = df.groupby("class").agg({
    "age": ["mean", "median"],
    "fare": ["mean", "max"]
})

print(agg_stats)
```
