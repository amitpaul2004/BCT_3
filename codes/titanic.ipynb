```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")

df.columns = df.columns.str.strip()

sorted_fare = df.sort_values("fare", ascending=False)

print(sorted_fare.head())
```

```python
# Set 'who' column as index
df_indexed = df.set_index("who")

print(df_indexed.head())
```
