```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\testing\BCT_3\data\social_media_engagement_dataset.csv")
```

```python
print(df.head())
```

```python
platform_counts = df['Platform'].value_counts()

print("\nPlatform Usage:\n")
print(platform_counts)
```

```python
most_used = platform_counts.idxmax()
print("\nMost Used Social Media Platform:", most_used)

# Plot
platform_counts.plot(kind='bar')
plt.title("Most Used Social Media Platform")
plt.xlabel("Platform")
plt.ylabel("Number of Posts")
plt.show()
```
