```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r'D:\testing\BCT\data\Road_Accidents_2017-Tables_3.7.csv'
df = pd.read_csv(path)
```

```python
df_clean = df[df['Weather condition'] != 'Total'].copy()
```

```python
df_clean['Fatality_Rate'] = (df_clean['Persons killed'] / df_clean['Number of Accidents']) * 100

# 4. Visualization
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_clean, x='Weather condition', y='Number of Accidents', palette='viridis')
plt.title('Accidents by Weather Condition (2017 Data)')
plt.show()
```
