```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Alice", "Charlie"],
    "Age": [25, 30, 25, 35]
}

df_dup = pd.DataFrame(data)

print("Before removing duplicates:")
print(df_dup)
```

```python
df_no_dup = df_dup.drop_duplicates()

print("After removing duplicates:")
print(df_no_dup)
```
